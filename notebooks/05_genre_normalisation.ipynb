```python
books["display_genre"] = (
    books["categories"]
    .map(genre_mapping)
    .fillna("Other")
)

books["display_genre"].value_counts()
```

```python
def assign_fantasy_genre(row):

    category = str(row["categories"]).lower()
    description = str(row["description"]).lower()

    fantasy_keywords = [
        "magic",
        "wizard",
        "witch",
        "dragon",
        "dragons",
        "spell",
        "sorcerer",
        "hogwarts",
        "elf",
        "elves",
        "dwarf",
        "orc",
        "kingdom",
        "fairy",
        "mythical",
        "quest"
    ]

    # Already Fantasy
    if category == "fantasy":
        return "Fantasy"

    # Description contains fantasy keywords
    if any(keyword in description for keyword in fantasy_keywords):
        return "Fantasy"

    return row["display_genre"]
```

```python
books["display_genre"] = books.apply(
    assign_fantasy_genre,
    axis=1
)
```
